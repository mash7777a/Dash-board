```python
from dash import Dash, html, dcc
```

```python
import plotly.express as px
import pandas as pd
from dash.dependencies import Input, Output
```

```python


app = Dash(__name__)

df = pd.read_csv('crimedata.csv')


app.layout = html.Div(children=[
#обложка
    html.Div(children=[
        html.H1(children='Analysis crime dataset', style={
            'text-align': 'center', 'color': '#FFFFFF', 'font-size': '48px', 'font-weight': 'bold', 'margin': '0 auto'}),
        html.P(children='in USA', style={
           'text-align': 'center', 'color': '#CFCFCF', 'max-width': '384px', 'margin': '4px auto'})
    ], style={
        'background-color': '#222222', 'height': '256px', 'display': 'flex', 'flex-direction': 'column', 'justify-content': 'center'}),

# 1й график
    html.Div([
        html.H3(children='Crimes in all states'),
        dcc.Dropdown(
            id='dropdown1',
            options=[{'value': 'arsonsPerPop', 'label': 'Arson'},
                     {'value': 'assaultPerPop', 'label': 'Assaults'},
                     {'value': 'autoTheftPerPop', 'label': 'Auto theft'},
                     {'value': 'burglPerPop', 'label': 'Burglaries'},
                     {'value': 'larcPerPop', 'label': 'Larcenies'},
                     {'value': 'murdPerPop', 'label': 'Murders'},
                     {'value': 'rapesPerPop', 'label': 'Rapes'},
                     {'value': 'robbbPerPop', 'label': 'Robberies'}
                     ],
            value='arsonsPerPop',
            clearable=False,
    ),
    dcc.Graph(id='output_graph1'),
    html.H4(
        children='We can do several conclusions, for example, larcenie is the most popular crime.',
        style={'text-align': 'start', 'font-weight': 'normal'}
    ),
    html.H4(
        children='!Important! In this data analysis, there are no states with no general data about, they dont make statistics',
        style={'text-decoration-line': 'underline', 'text-decoration-color': 'DarkSeaGreen'}
    )], style={'background-color': '#F8F8FF', 'display': 'flex', 'flex-direction': 'column',
                   'justify-content': 'center'}),

# 2й график
    html.Div([
        html.Br(),
        html.H3(children='Crimes and races'),

        dcc.Dropdown(
            id='dropdown2',
            options=[{'value': 'arsonsPerPop', 'label': 'Arson'},
                     {'value': 'assaultPerPop', 'label': 'Assaults'},
                     {'value': 'autoTheftPerPop', 'label': 'Auto theft'},
                     {'value': 'burglPerPop', 'label': 'Burglaries'},
                     {'value': 'larcPerPop', 'label': 'Larcenies'},
                     {'value': 'murdPerPop', 'label': 'Murders'},
                     {'value': 'rapesPerPop', 'label': 'Rapes'},
                     {'value': 'robbbPerPop', 'label': 'Robberies'}
                     ],
            value='arsonsPerPop',
            clearable=False,

        ),
        html.Br(),
        dcc.RadioItems(
            id='radio-items2',
            options=[{'value': 'racepctblack', 'label': 'Blacks'},
                     {'value': 'racePctWhite', 'label': 'Whites'},
                     {'value': 'racePctAsian', 'label': 'Asians'},
                     {'value': 'racePctHisp', 'label': 'Hispanics'}
                     ],
            value='racepctblack'
        ),
        dcc.Graph(id='output_graph2'),
        html.H4(
            children='Thanks to this schedule, we can now say that if your car is stolen, which color the stealer would be',
            style={'text-align': 'start', 'font-weight': 'normal'}
        )], style={'background-color': '#F8F8FF', 'display': 'flex', 'flex-direction': 'column',
                   'justify-content': 'center'}),

# 3й график
    html.Div([
        html.Br(),
        html.H3(children='People without places of residence'),

        html.Div([
            dcc.RadioItems(
            id='radio-items3',
            options=[{'value': 'NumStreet', 'label': 'Streets'},
                     {'value': 'NumInShelters', 'label': 'Shelters'}
                     ],
            value='NumStreet'
        ),
            dcc.Graph(id='output_graph3',
                      style={'width': '50%'}),
            dcc.Graph(id='output_graph3_1',
                      style={'width': '50%', 'position': 'relative', 'top': '-450px', 'left': '620px'})
        ]),

        dcc.Checklist(id="checklist3",
                      options=df.state.unique(),
                      value=df.state.unique()[39:48],
                      inline=True,
                      style={'position': 'relative', 'top': '-450px'}),

        html.H4(
            children='Interestingly, the graph shows that homeless people are usually not the ones who steal something from other homes',
            style={'text-align': 'start', 'font-weight': 'normal', 'position': 'relative', 'top': '-450px'}
        )], style={'background-color': '#F8F8FF', 'display': 'flex', 'flex-direction': 'column',
                       'justify-content': 'center'}),

# 4й график
    html.Div([
        html.Br(),
        html.H3(children='Murders and different factors'),

        html.Div([
            dcc.Dropdown(
                id='dropdown4',
                options=[{'value': 'PopDens', 'label': 'population density'},
                         {'value': 'PolicPerPop', 'label': 'police number'},
                         {'value': 'NumImmig', 'label': 'immigrants number'}
                         ],
                value='PopDens',
                clearable=False,
                multi=True
            ),
        dcc.Graph(id='output_graph4'),

        dcc.Checklist(id="checklist4",
                      options=df.state.unique(),
                      value=df.state.unique()[39:48],
                      inline=True),

        html.H4(
            children='Finally, there is no general relationship between the number of police officers, migrants, population density and thefts. Somewhere the crime rate is higher, but less police and vice versa ',
            style={'text-align': 'start', 'font-weight': 'normal'}
        )], style={'background-color': '#F8F8FF', 'display': 'flex', 'flex-direction': 'column',
                   'justify-content': 'center'})],
        style={'position': 'relative', 'top': '-450px'}),
    html.Div([
        html.Br(),
        html.H3(children='В целом, это было интересное наблюдение за взаимосвязью различных факторов с преступностью, чаще всего не было общей динамики между штатами, '
                         'и, пожалуй, единственое, что однозначно повышало преступность, была бедность населения.')],
    style={'background-color': '#F8F8FF', 'display': 'flex', 'flex-direction': 'column',
                   'justify-content': 'center', 'position': 'relative', 'top': '-450px'})

])


@app.callback(
    Output('output_graph1', 'figure'),
    Input('dropdown1', 'value')
)
def update_output(value):
    crimes = df.groupby("state").agg({value: 'mean'}).reset_index()
    crimes = crimes.dropna()
    figure = px.bar(crimes, x='state', y=str(value), labels={'state': 'States', str(value): 'Per Population'},
                    color_discrete_sequence=['SteelBlue'], template='seaborn')
    return figure


@app.callback(
    Output('output_graph2', 'figure'),
    Input('radio-items2', 'value'),
    Input('dropdown2', 'value'),
)
def update_output2(value_drop, value_radio):
    crimes = df.groupby("state").agg({value_drop: 'mean', value_radio: 'mean'}).reset_index()
    crimes = crimes.dropna()
    figure = px.bar(crimes, x='state', y=[value_drop, value_radio],
                    labels={'state': 'States', 'value': 'Per population', 'variable': 'colors'},
                    color_discrete_sequence=['Teal', 'DarkSeaGreen'], barmode='group', template="seaborn")
    return figure


@app.callback(
    Output('output_graph3', 'figure'),
    Output('output_graph3_1', 'figure'),
    Input('radio-items3', 'value'),
    Input('checklist3', 'value')
)
def update_output3(value_radio, value_check):
    crimes = df.groupby('state').agg({'burglaries': 'mean', str(value_radio): 'mean'}).reset_index()
    crimes = crimes[crimes['state'].isin(value_check)]
    crimes = crimes.dropna()
    if value_radio == 'NumStreet':
        label = 'Street'
    else:
        label = 'Shelter'
    figure1 = px.bar(crimes, x='state', y=[str(value_radio)],
                    labels={'state': 'States', 'value': 'amount of people'},
                    color_discrete_sequence=['DarkSeaGreen'],
                    template="seaborn")
    figure2 = px.bar(crimes, x='state', y=[str(value_radio), 'burglaries'],
                     labels={'state': 'States', 'value': 'amount of people', 'variable': 'colors'},
                     barmode='group',
                     category_orders={'variable': [label, 'burglaries']},
                     color_discrete_map={label: "DarkSeaGreen", "burglaries": "SteelBlue"},
                     template="seaborn")
    return figure1, figure2

@app.callback(
    Output('output_graph4', 'figure'),
    Input('dropdown4', 'value'),
    Input('checklist4', 'value')
)
def update_output4(value_drop, value_check):
    crimes = df.groupby('state').agg(
        {'PolicPerPop': 'mean', 'PopDens': 'mean', 'NumImmig': 'sum', 'population': 'sum', 'larcPerPop': 'mean'}).reset_index()
    crimes['NumImmig'] = (crimes['NumImmig'] / crimes['population']) * 100000
    crimes = crimes[crimes['state'].isin(value_check)]
    crimes = crimes.dropna()
    value_drop = [i for i in value_drop]
    value_drop.append('larcPerPop')

    figure = px.bar(crimes, x='state', y=value_drop,
                    labels={'state': 'States', 'value': 'per 100k population', 'variable': 'colors'},
                    color_discrete_map={
                        'PopDens': "DarkSeaGreen", "NumImmig": "SteelBlue",
                        'PolicPerPop': 'MediumPurple', 'larcPerPop': 'DarkSlateGray'},
                    barmode='group',
                    template="seaborn")
    return figure

if __name__ == '__main__':
    app.run_server(debug=True)


```
